```python
def all_results_from(pair):
    '''
    Yield all of the possible positive integer results
    assume b > a
    '''
    a,b = pair
    # Addition and multiplication are commutative.
    yield '+', a + b
    yield 'x', a * b
    # Subtraction and division are not commutative.
    c = b - a
    if c > 0:
        yield '-', c
        
    if b % a == 0:
        yield '/', b // a
```

```python

from collections import defaultdict
from operator import mul
from functools import reduce

def solve(numbers, target):
    #initialisation before doing the loop
    graph = {}
    #normalise/sort
    numbers = tuple(sorted(numbers))
    
    #caches this (intelligently)
    pairs_cache = [[], [],[(0,1)]]
    for k in range(2,len(numbers) + 1):
        L = pairs_cache[-1][:]
        L.extend([(i,k) for i in range(k)])
        pairs_cache.append(L)
     
    def next_node(numbers):
        #careful this is going to be recursive
        if len(numbers) == 1: return
    
        #the biggest number u can make is (usually) the product
        if min(numbers) > 1 and reduce(mul,numbers) < target: return
    
        for i,j in pairs_cache[len(numbers)]:
            pair = numbers[i], numbers[j]
            #should cache this too ?
            rejects = [x for k,x in enumerate(numbers) 
                          if k not in [i,j] ]
            
            for op, c in all_results_from(pair):
                node = tuple(sorted(rejects + [c]))
                #print('%d %s %d'%(pair[1], op, pair[0]), node)
                if node not in graph:
                    #initialise node
                    graph[node] = defaultdict(set)
                graph[node][numbers].add((pair, op, node))
                #bail out because u won
                if c == target: return
                next_node(node)
                
    next_node(numbers)
    return graph

def find_path(graph, 
              start, end, 
              path=[]):
    
        path = path + [start]
        if start == end:
            print(path)
            return path
        if start not in graph: return None
        
        for node in graph[start]:
            if node not in path:
                newpath = find_path(graph, node, end, path) 
        return None
```
